```python
## Question 2

import time

#Matrix Multiplication using List

S1 = time.time()

M1 = [[1,2,3],
      [2,3,4],
      [5,6,7]]

M2 = [[2,2,3],
      [4,0,4],
      [8,1,7]]

M = [[0 for _ in range(3)] for _ in range(3)]

#Multiplying it using for loop

temp = 0

for i in range(3):
  for j in range(3):
    temp=0
    for k in range(3):
      temp = temp + M1[i][k]*M2[k][j]
    M[i][j]=temp

S2 = time.time()

print(M)

S4 = time.time()


#Multiply using numpy


import numpy as np

M1 = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [5, 6, 7]
])

M2 = np.array([
    [2, 2, 3],
    [4, 0, 4],
    [8, 1, 7]
])

result_matrix = np.dot(M1, M2)

S3 = time.time()

print(result_matrix)



print("time for traditional method: ",S2-S1,"time using numpy array",S3-S4)

# We can clearly observe that numpy is faster then traditional implementation because of its implementation in C and optimized for numerical operations. NumPy uses highly optimized, pre-compiled functions that take advantage of low-level optimizations, making it more efficient than the pure Python list of lists approach.

```

```python
#time complexity of finding the median using given method is O(n^2)

import time
import numpy as np

def find_median(arr):
    n = len(arr)
    if n % 2 == 0:
        mid1 = find_kth_largest(arr, n // 2)
        mid2 = find_kth_largest(arr, n // 2 - 1)
        return (mid1 + mid2) / 2
    else:
        return find_kth_largest(arr, n // 2)

def find_kth_largest(arr, k):
    for i in range(k + 1):
        max_index = find_max_index(arr)
        if i == k:
            return arr[max_index]
        arr[max_index] = float('-inf')

def find_max_index(arr):
    max_index = 0
    for i in range(1, len(arr)):
        if arr[i] > arr[max_index]:
            max_index = i
    return max_index

# Python3 program of Quick Select (I have refered from Geeks for Geeks)
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_select(arr, low, high, k):
    if low <= high:
        pivot_index = partition(arr, low, high)

        if pivot_index == k:
            return arr[pivot_index]
        elif pivot_index < k:
            return quick_select(arr, pivot_index + 1, high, k)
        else:
            return quick_select(arr, low, pivot_index - 1, k)

def find_median_quick(arr):
    n = len(arr)
    if n % 2 == 0:
        median1 = quick_select(arr, 0, n - 1, n // 2 - 1)
        median2 = quick_select(arr, 0, n - 1, n // 2)
        return (median1 + median2) / 2
    else:
        return quick_select(arr, 0, n - 1, n // 2)




arr1 = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
arr2 = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
arrnp = np.array(arr1)

S1 = time.time()
median1 = find_median(arr1)
S2 = time.time()
median2 = find_median_quick(arr2)
S3 = time.time()
median3 = np.median(arrnp)
S4 = time.time()

print("Median Worst:", median1,"Median Quick:", median2,"Median Numpy", median3)
print("Time Median Worst:", S2-S1,"Time Median Quick:", S3-S2,"Time Median Numpy", S4-S3)
```

```python
import jax
import jax.numpy as np
from jax import grad, random

key = random.PRNGKey(42)

def fun(x, y):
    return ((x**2) * y) + ((y**3) * np.sin(x))

def sol_x(x,y):
    return (2*x*y) + ((y**3)*np.cos(x))

def sol_y(x,y):
    return (x**2) + ((3*y**2)*np.sin(x))

# Define random values for x and y
x_value = random.uniform(key)
y_value = random.uniform(key)

# Computing the gradient for both values
grad_fun_x = grad(fun, argnums=0)(x_value, y_value)
grad_fun_y = grad(fun, argnums=1)(x_value, y_value)

#print both the values
print("Gradent wrt to x: ",grad_fun_x,"Calculated values: ",sol_x(x_value,y_value))
print("Gradent wrt to x: ",grad_fun_y,"Calculated values: ",sol_y(x_value,y_value))

```

```python
class Student:
    def __init__(self, roll_number, name, marks=None):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks if marks else {}

class Branch:
    def __init__(self, name, students=None):
        self.name = name
        self.students = students if students else []

    def add_student(self, student):
        self.students.append(student)

class AcademicYear:
    def __init__(self, year, branches=None):
        self.year = year
        self.branches = branches if branches else []

    def add_branch(self, branch):
        self.branches.append(branch)

# Create instances representing the data structure
student_1 = Student(1, 'Zaqi', {'Maths': 100, 'English': 70})
student_2 = Student(2, 'Mithil', {'Maths': 90, 'English': 80})
branch_1 = Branch('Branch 1', [student_1, student_2])

student_3 = Student(3, 'Abhishek', {'Maths': 95, 'English': 85})
student_4 = Student(4, 'Mihir', {'Maths': 85, 'English': 75})
branch_2 = Branch('Branch 2', [student_3, student_4])

student_5 = Student(1, 'Atharva', {'Maths': 100, 'English': 70})
student_6 = Student(2, 'Mithil', {'Maths': 90, 'English': 80})
branch_3 = Branch('Branch 1', [student_5, student_6])

student_7 = Student(3, 'Kunj', {'Maths': 15, 'English': 5})
student_8 = Student(4, 'Mihir', {'Maths': 85, 'English': 75})
branch_4 = Branch('Branch 2', [student_7, student_8])

year_2022 = AcademicYear(2022, [branch_1, branch_2])
year_2023 = AcademicYear(2023, [branch_3, branch_4])

# Create the overall database
database = [year_2022,year_2023]

# Accessing specific information
print(database[0].branches[0].students[0].name)
print(database[1].branches[1].students[0].marks['English'])
```

```python
import numpy as np

def custom_argmin(array):

    #store shape of array
    shape = array.shape

    #flatten array
    array = array.flatten()

    #find index of minimum value using for loop
    min_index = 0
    for i in range(len(array)):
        if array[i] < array[min_index]:
            min_index = i

    #return index of minimum value in original array shape
    return np.unravel_index(min_index, shape)

# Test the function
arr = np.array([[0, 1, 4], [1, 5, 9], [2, 6, -1]])

# Use the custom function
index_custom = custom_argmin(arr)

# Use np.argmin to verify
index_np = np.unravel_index(np.argmin(arr), arr.shape)

print("Index computed by custom function:", index_custom)
print("Index computed by np.argmin:", index_np)
```
